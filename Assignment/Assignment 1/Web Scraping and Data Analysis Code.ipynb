```python
def scrapWeb(url):
    browser = webdriver.Chrome()
    browser.get(url)

    # wait until the page loads
    wait = WebDriverWait(browser, 10)
    wait.until_not(EC.presence_of_element_located((By.CLASS_NAME, "grid")))

    soup = BeautifulSoup(browser.page_source, "html.parser")
    browser.close()
    return soup
```

```python
moives = []
for page in range(0, 100, 10):
    url = 'https://maoyan.com/board/4?offset=' + str(page)
    result = scrapWeb(url)
    print(result)
    pattern = re.compile('<dd>.*?board-index.*?>(\d+)</i>.*?name.*?href="(.*?)".*?title="(.*?)".*?star'
                         +'.*?主演：(.*?)</p>.*?releasetime.*?上映时间：(\d{4}-\d{1,2}-\d{1,2}|\d{4})'
                         +'.*?</p>.*?"integer">(.*?)</i>.*?fraction">(.*?)</i>.*?</dd>', re.S)
    contents = re.findall(pattern, str(result))
#     print(contents)
    moives += contents
    print(moives)
    time.sleep(1)
```

```python
data_list = []
for i in moives:
    sin_list = []
    for j in range(len(i)-1):
        if j == 3:
            sin_list.append(i[j].replace("\n","").strip())
        elif j == 5:
            sin_list.append(i[j]+i[j+1])
        else:
            sin_list.append(i[j])
    data_list.append(sin_list)
print(data_list)
```

```python
def scrapSubWeb(link):
    url = 'https://maoyan.com' + link[1]
    result = scrapWeb(url)
    
    match = result.find_all(class_="ellipsis")
    findType = re.compile('> (.*) <')
    MovieType = re.findall(findType, str(match[1]))
    print(MovieType)
    strType = ''
    for i in MovieType:
        strType += i + '，'
    link.append(strType[:-1])
    
    findDuration = re.compile('(...分钟)')
    duration = re.findall(findDuration, str(match[2]))
    print(duration)
    link.append(duration[0])
    
    findRegion = re.compile('([\u4e00-\u9fa5]{0,20})\n.*\/')
    region = re.findall(findRegion, str(match[2]))
    print(region[0])
    link.append(region[0])

    matchDirector = result.find_all(class_="celebrity-group")[0]
    matchDirector = matchDirector.find_all(class_="name")
    findDirector = re.compile('<a.*?>(.*?)</a>', re.S)
    director = re.findall(findDirector, str(matchDirector))
    print(director)
    strDirector = ''
    for i in director:
        strDirector += i.replace("\n","").strip() + '，'
    link.append(strDirector[:-1])
    
    findIncome = re.compile('>(.*)<')
    try:
        matchDirector = str(result.find_all(class_="mbox-name")[2])
        income = str(re.findall(findIncome, matchDirector)[0])
        if income != '暂无':
            income += '万'
    except:
        income = '暂无'    
    print(income)
    link.append(income)
    
    time.sleep(2)
```

```python
# Region (Country) distribution

plt.figure(figsize=(10,10))
plt.rcParams['font.sans-serif']=['SimHei']
country = data.groupby('Region')['Region'].count()

patches,l_text,p_text = plt.pie(list(country), labels=list(country.index), autopct='%1.1f%%', counterclock=False)
for t in p_text:
    t.set_size(15)

for t in l_text:
    t.set_size(20)
    
plt.axis('equal') 
plt.title('Region/Country distribution')
plt.savefig("image\\region distribution.png", dpi=200, bbox_inches='tight')  
```

```python
# Movie Type Word Cloud

from pyecharts.globals import CurrentConfig, NotebookType  
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK

from pyecharts import options as opts
from pyecharts.charts import WordCloud

type_list = []
for i in data['Type']:
    if '，' in i:
        temp = i.split('，')
        type_list += temp
    else:
        type_list.append(i)
# print(type_list, len(type_list))

type_count = Counter(type_list)
# print (dict(type_count))

name = []
value = []
for i in type_count.keys():
    name.append(i)
    value.append(type_count[i])
# print(name, value)

words=[(i,j) for i ,j in zip(name,value)]
worldcloud=(
     WordCloud()
     .add("", words, word_size_range=[20, 200])
     .set_global_opts(title_opts=opts.TitleOpts(title="Movie Type Word Cloud"))
     )

worldcloud.render_notebook()
```

```python
# Release year distribution of movies

date_list = []
for i in data['Release date']:
    if '/' in i:
        temp = i.split('/')[0]
        date_list.append(int(temp))
    else:
        date_list.append(int(i))
# print(date_list, len(date_list))

date_count = Counter(date_list)
# print (dict(date_count))

date = []
count = []
for i in sorted(date_count.keys()):
    date.append(i)
    count.append(date_count[i])
# print(date, count)

plt.figure(figsize=(15, 10), dpi=80)

plt.plot(date, count)
plt.title('Release year statistics of movies', fontsize=20)
plt.xlabel('Release Year', fontsize=15)
plt.grid(alpha=0.9)
plt.savefig("image\\Release year statistics of movies.png", dpi=200, bbox_inches='tight')
```

```python
# Cumulative income of directors

director_income = data[['Name of director', 'Cumulative income']]
# print(director_income['Name of director'])

director_dict = {}

for indexs in director_income.index:
    name, value = director_income.loc[indexs].values
    if value == '暂无':
        continue
    else:
        if '，' in name:
            name = name.split('，')
            for i in name:
                if i in director_dict:
                    director_dict[i] = director_dict[i] + int(value[:-1])
                else:
                    director_dict[i] = int(value[:-1])
        else:
            if name in director_dict:
                director_dict[name] = director_dict[name] + int(value[:-1])
            else:
                director_dict[name] = int(value[:-1])

people = []
performance = []
for i in director_dict.keys():
    people.append(i)
    performance.append(director_dict[i])
# print(people, performance)

plt.rcdefaults()
fig, ax = plt.subplots(figsize=(16, 20))

plt.rcParams['font.sans-serif']=['SimHei']

y_pos = np.arange(len(people))

total_width, n = 0.8, 2
width = total_width / n
y_pos=y_pos - (total_width - width) / 2

b = ax.barh(y_pos, performance, align='center', color='green', ecolor='black', height=0.6)
for rect in b:
    w=rect.get_width()
    ax.text(w,rect.get_y()+rect.get_height()/2,'%1.1f'%w,ha='left',va='center')

ax.set_yticks(y_pos+width/2.0)
ax.set_yticklabels(people)
ax.invert_yaxis()
ax.set_xlabel('Cumulative Income (million yuan)', fontsize=20)
ax.set_title('Cumulative income of directors', fontsize=15)
plt.savefig("image\\Cumulative income of directors.png", dpi=200, bbox_inches='tight')
```

```python
# Duration distribution of movies

time = []
for i in data['Duration']:
    time.append(int(i[:-2]))

plt.hist(time, range(min(time), max(time) + 10, 10), density=True, color="#33ffff")

plt.xticks(range(min(time), max(time) + 10, 10))
plt.grid(alpha=0.4)

plt.title('Duration distribution of movies', fontsize=20)
plt.xlabel('Duration', fontsize=15)
plt.ylabel('Proportion', fontsize=15)
plt.savefig("image\\Duration distribution of movies.png", dpi=200, bbox_inches='tight')
```

```python
# Relationship between rating, type, and income

types = type_count.keys()
type_convert = {}
type_value = 0
for i in types:
    type_convert[i] = type_value
    type_value += 1
# print(type_convert)

dataset = data[['Rating', 'Type', 'Cumulative income']]
Point = []

for indexs in dataset.index:
    x, y, z = dataset.loc[indexs].values
    if z == '暂无':
        continue
    else:
        if '，' in y:
            for i in y.split('，'):
                Point.append([float(x), type_convert[i], int(z[:-1])])
        else:
            Point.append([float(x), type_convert[y], int(z[:-1])])
# print(Point)
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']

fig = plt.figure(figsize=(20, 20))
ax = fig.gca(projection='3d')

X, Y, Z = [a[0] for a in Point], [b[1] for b in Point], [c[2] for c in Point]
# print(X, Y, Z, len(X), len(Y), len(Z))

ax.scatter(X, Y, Z, s=50, marker='o', color='#FF2D2D')
plt.yticks(range(len(types)), types)

plt.title('Relationship between rating, type, and cumulative income', fontsize=20)
# plt.xlabel('Rating', fontsize=15)
# plt.ylabel('Type', fontsize=15, rotation=50)
plt.savefig("image\\Relationship between rating, type, and cumulative income.png", dpi=200, bbox_inches='tight')
```

```python
# Relationship between rank, rating, and income

ds = data[['Rank', 'Rating', 'Cumulative income']]
Point = []

for indexs in ds.index:
    x, y, z = ds.loc[indexs].values
    if z == '暂无':
        continue
    else:
        Point.append([int(x), float(y), int(z[:-1])])
# print(Point)

fig = plt.figure(figsize=(15, 15))

# first figure
ax = fig.add_subplot(1, 2, 1, projection='3d')

X, Y, Z = [a[1] for a in Point], [b[0] for b in Point], [c[2] for c in Point]
# print(X, Y, Z, len(X), len(Y), len(Z))

ax.scatter3D(X, Y, Z, cmap='Blues')
ax.plot3D(X, Y, Z, 'gray')

plt.title('Oiginal Figure', fontsize=20)
plt.xlabel('Rating', fontsize=15)
plt.ylabel('Rank', fontsize=15, rotation=50)

# second figure
ax = fig.add_subplot(1, 2, 2, projection='3d')

x_z = [min(X)-0.5]*len(X)
y_z = [max(Y)+0.5]*len(Y)
x_y = [min(Z)-0.5]*len(Z)

ax.plot(x_z, Y, Z)
ax.plot(X, y_z, Z)
ax.plot(X, Y, x_y)

plt.title('Projection Figure', fontsize=20)
plt.xlabel('Rating', fontsize=15)
plt.ylabel('Rank', fontsize=15, rotation=50)

plt.savefig("image\\Relationship between rank, rating, and income.png", dpi=200, bbox_inches='tight')
```
