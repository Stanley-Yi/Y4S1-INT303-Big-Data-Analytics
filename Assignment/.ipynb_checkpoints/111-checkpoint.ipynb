```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

datas = pd.read_csv('XingboWei-1824150.csv', encoding='utf-8')

long_time = list(datas['时长'])

d = 10
num_bins = int((max(long_time) - min(long_time)) / d)

plt.hist(long_time, range(min(long_time), max(long_time) + d, d), density=True, color="#33ff44")

plt.xticks(range(min(long_time), max(long_time) + d, d))
plt.grid(alpha=0.4)

plt.title('The Movie Duration Distribution')
plt.xlabel('Duration')
plt.ylabel('Proportion')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter

plt.rcParams['font.sans-serif']=['SimHei']

datas = pd.read_csv('XingboWei-1824150.csv', encoding='utf-8')
s = ''
for i in range(99):
    s += datas.iloc[i, 2]+','
s += datas.iloc[99, 1]
# print(s)
authors = s.split(',')
# print(authors)
c = Counter(authors)
# print(c)
items = c.most_common(6)
# print(items)

author = []
count = []
for item in items:
    author.append(item[0])
    count.append(item[1])

# print(author)
# print(count)
plt.bar(author, count, color=['#33ff00', '#33ff33', '#33ff66', '#33ff99', '#33ffcc', '#33ffff'])
plt.title('Occurrences Order For Stars')
plt.xlabel('Actor')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

datas = pd.read_csv('XingboWei-1824150.csv', encoding='utf-8')

datas['year'] = datas['上映时间']


year = datas.groupby('year')['year'].count()
plt.figure(figsize=(20, 10), dpi=80)

plt.plot(list(year.index), list(year))
plt.title('电影年份的分布情况')
plt.xlabel('年份')
plt.ylabel('电影数量')
plt.grid(alpha=0.9)
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter

from wordcloud import WordCloud

plt.rcParams['font.sans-serif']=['SimHei']

datas = pd.read_csv('XingboWei-1824150.csv', encoding='utf-8')
s = ''
for i in range(99):
    s += datas.iloc[i, 5]+','
s += datas.iloc[99, 1]
# print(s)
authors = s.split(',')
# print(authors)
c = Counter(authors)
# print(c)
items = c.most_common(6)
print(items)
str = ""
for i in items:
    for ii in range(i[1]):
        str = str + i[0] + " "
wc = WordCloud(
        # 设置字体，不指定就会出现乱码
        # 设置背景色
        background_color='white',
        # 设置背景宽
        width=500,
        # 设置背景高
        height=350,
        # 最大字体
        max_font_size=50,
        # 最小字体
        min_font_size=10,
        mode='RGBA'
        #colormap='pink'
        )
# 产生词云
wc.generate(str)
# 保存图片
# 4.显示图片
# 指定所绘图名称
plt.figure("jay")
# 以图片的形式显示词云
plt.imshow(wc)
# 关闭图像坐标系
plt.axis("off")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
from collections import Counter
datas = pd.read_csv('XingboWei-1824150.csv', encoding='utf-8')
s = ''
for i in range(99):
    s += datas.iloc[i, 5]+','
s += datas.iloc[99, 1]
authors = s.split(',')
c = Counter(authors)
items = c.most_common(6)
#print(items)
array = []
for i in items:
    array.append(i[1])
np.array(array)

matplotlib.rcParams['font.family']='SimHei'
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
labels = np.array(['剧情', '爱情', '喜剧', '犯罪', '冒险', '奇幻'])
nAttr = 6

data = array
angles = np.linspace(0, 2*np.pi, nAttr, endpoint=False)
data = np.concatenate((data, [data[0]]))
angles = np.concatenate((angles, [angles[0]]))
labels=np.concatenate((labels, [labels[0]]))
fig = plt.figure(facecolor="white")

plt.subplot(111, polar=True)
plt.plot(angles, data, color ='g', linewidth=2)
plt.fill(angles, data, facecolor='g', alpha=0.25)
plt.thetagrids(angles*180/np.pi, labels)
plt.figtext(0.52, 0.95, 'Radar for Type', ha='center')
plt.grid(True)
plt.show()
```
